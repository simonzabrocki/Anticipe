```python
import pandas as pd 
from index.utils import ISO_to_Everything, ISO_to_Country
```

```python
data = pd.read_csv('data/indicator/GJ2/processed/GJ2_SDG.csv')
```

```python
missing_ISO_high_upper_income = ISO_to_Everything(data[['ISO', 'Indicator']].drop_duplicates()).reset_index()
missing_ISO_high_upper_income = missing_ISO_high_upper_income[missing_ISO_high_upper_income.Indicator.isna()].query('IncomeLevel in ["High income"]')[['ISO', 'Country']]#.ISO
```

```python
from processing.imputation import create_full_index
```

```python
high_inc_df = create_full_index(missing_ISO_high_upper_income.assign(Year=2020)).assign(Value=0, Imputed=True, Corrected=False, Indicator='GJ2', From='SDG API')
high_inc_df['Country'] = ISO_to_Country(high_inc_df.ISO)
high_inc_df['Source'] = data.Source.unique()[0]
high_inc_df['URL'] = data.URL.unique()[0]
high_inc_df['DownloadDate'] = data.DownloadDate.unique()[0]
high_inc_df['Description'] = data.Description.unique()[0]
```

```python
df = pd.concat([data, high_inc_df])
df.to_csv('data/indicator/GJ2/processed/GJ2_SDGbis.csv', index=False)
```
