```python
from ggdata import download

from ggdata.downloaders.downloader import SDG_Downloader

Downloader = SDG_Downloader('https://unstats.un.org/SDGAPI/v1/sdg/Series/Data') 

# Above-ground biomass stock in forest (tonnes per hectare) 
parameter1 = {'seriesCode': 'AG_LND_FRSTBIOPHA' }
data1 = Downloader.download_data('example1.json', parameter1) # Save the data as a JSON file

# Proportion of safely treated domestic wastewater flows (%) 
parameter1 = {'seriesCode': 'EN_WWT_WWDS' }
new_data1 = Downloader.download_data('New_indicator1.json', parameter1) # Save the data as a JSON file


# Forest area annual net change rate (%) 
parameter2 = {'seriesCode': 'AG_LND_FRSTCHG' }
data2 = Downloader.download_data('example2.json', parameter2) # Save the data as a JSON file

# Forest area under an independently verified forest management certification scheme (thousands of hectares) 
parameter3 = {'seriesCode': 'AG_LND_FRSTCERT' }
data3 = Downloader.download_data('example3.json', parameter3) # Save the data as a JSON file

# Proportion of forest area under a long-term management plan (%) 
parameter4 = {'seriesCode': 'AG_LND_FRSTMGT' }
data4 = Downloader.download_data('example4.json', parameter4) # Save the data as a JSON file

# Proportion of forest area within legally established protected areas (%) 
parameter5 = {'seriesCode': 'AG_LND_FRSTPRCT' }
data5 = Downloader.download_data('example5.json', parameter5) # Save the data as a JSON file

# water 
parameter5 = {'seriesCode': 'AB1.0_SDG' }
data6 = Downloader.download_data('example5.json', parameter5) # Save the data as a JSON file
```

```python
from ggdata.preprocessors.SDG import SDG_Preprocessor

Preprocessor = SDG_Preprocessor('test') #file argument to change (Used to preprocess special cases)
information = {'Variable': 'Test', 'From': 'SDG API'} # Let's you add more information to the dataframe
df1 = Preprocessor.preprocess(data1, information)
df1.to_excel("Above-ground biomass stock in forest.xlsx")

#New indicator 

new1 = Preprocessor.preprocess(new_data1, information)
new1.to_excel("Proportion of safely treated domestic wastewater flows.xlsx")



df2 = Preprocessor.preprocess(data2, information)
df2.to_excel("Forest area annual net change rate.xlsx")

df3 = Preprocessor.preprocess(data3, information)
df3.to_excel("Forest area under an independently verified forest management certification scheme.xlsx")

df4 = Preprocessor.preprocess(data4, information)
df4.to_excel("Proportion of forest area under a long-term management plan.xlsx")

df5 = Preprocessor.preprocess(data5, information)
df5.to_excel("Proportion of forest area within legally established protected areas.xlsx")

df6 = Preprocessor.preprocess(data5, information)
df6.to_excel("water.xlsx")
```
