```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def preprocess_bis():
    all_data = pd.read_excel("data/indicator/EE3/raw/International_LPI.M.xlsx", skiprows=2 , sheet_name=[0,1,2,3,4,5])
    df2 = pd.DataFrame()
    year = [2018, 2016, 2014, 2012, 2010, 2007]
    for i in all_data.keys():
        df = all_data[i]
        df['Year'] = year[i]
        df2 = df2.append(df)
    df2.rename(columns = {'Code':'ISO', 'score': 'Value'}, inplace = True)
    df2[['ISO', 'Value']]
    df2 = df2[['ISO', 'Value' ,'Year']]
    df2.reset_index(drop=True, inplace=True)
    return df2

config =  {'Variable': 'EE3',
              'function': preprocess_bis,
              'Description': 'Efficiency in sustainable transport',
              'Source': 'logistics performance index (LPI) of the World Bank ',
              'URL': 'https://lpi.worldbank.org/ '}
```
